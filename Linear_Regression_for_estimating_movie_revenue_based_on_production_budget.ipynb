```
import pandas as pd
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```
data  = pd.read_csv('movie_clean_data_.csv')
```

```
data.describe()
```

```
X = DataFrame(data, columns=['Production_budget'])
y = DataFrame(data, columns=['Worldwide_gross'])
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=1/3,random_state=0)
```

```
linreg = LinearRegression()
```

```
linreg.fit(X,y)
```

```
linreg.coef_
```

```
linreg.intercept_
```

```
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.scatter(X_test,y_test,alpha=0.3)
plt.plot(X_test,linreg.predict(X_test),color = 'red',linewidth = 3)
plt.xlabel("Production Budget")
plt.ylabel("Worldwide Gross")
plt.title("Revenue Estimation")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```
linreg.score(X_test,y_test)
```

```
linreg.score(X_train,y_train)
```
